qn 2 Python If else

```python
n = int(input())
if n % 2 == 1:
  print("Weird")
elif n % 2 == 0 and 2 <= n <= 5:
  print("Not Weird")
elif n % 2 == 0 and 6 <= n <= 20:
  print("Weird")
else:
  print("Not Weird")
```

qn 6 Write a function

```python
def is_leap(year):
  if(year % 400 == 0):
    return True
  if(year % 100 == 0):
    return False
  if(year % 4 == 0):
    return True
  return False
year = int(input().strip())
print(is_leap(year))
```

qn 7 Print Function

```python
n = int(input())
result = ""
for i in range(1, n+1):
  result += str(i)
print(result)
```
